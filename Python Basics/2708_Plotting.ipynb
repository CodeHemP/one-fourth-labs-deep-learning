```
import matplotlib.pyplot as plt
import numpy as np
```

```
x = np.random.random((10,1))
plt.plot(x)
plt.show()
```

```
plt.plot(x,'*')
plt.show()
```

```
y = np.power(x,0.5)
plt.plot(x,y)
plt.show()
```

```
x = np.linspace(0,10,100)
y = np.power(x,0.5)
plt.plot(x,y)
plt.show()
```

```
plt.plot(x,y,'*-')
plt.show()
```
